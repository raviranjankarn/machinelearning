```python
from tkinter import *
```

```python
coeff={'Cylinders':-0.389469
       ,'Displacement':0.021584
       ,'Horsepower':-0.012372
       ,'Weight':-0.007001
       ,'Acceleration':0.129544
       ,'Year':0.767744
       ,'Origin_Europe':0.833883
       ,'Origin_Japan':0.719824
       ,'Origin_US':-1.553706
       ,'Intercept':-16.203123
      }
```

```python
master = Tk()
master.title("Car MPG Simulator")
Cylinders_var = DoubleVar() 
Displacement_var = DoubleVar()
Horsepower_var = DoubleVar()
Weight_var = DoubleVar()
Acceleration_var = DoubleVar()
Year_var = DoubleVar()
MPG_var = DoubleVar()
Radio_var = IntVar()

def trace_method(*args):
    w_MPG.config(state="normal")
    w_MPG.set(w_Cylinder.get()*coeff['Cylinders']
            +w_Displacement.get()*coeff['Displacement']
            +w_Horsepower.get()*coeff['Horsepower']
            +w_Weight.get()*coeff['Weight']
            +w_Acceleration.get()*coeff['Acceleration']
            +(w_Year.get()-1900)*coeff['Year']
            +(coeff['Origin_Europe'] 
                 if Radio_var.get()==1 
                 else coeff['Origin_US']
                   if Radio_var.get()==2 
                   else coeff['Origin_Japan'] )
            +coeff['Intercept']
            )
    w_MPG.config(state="disabled")
    

    
w_Cylinder=Scale(master, from_=3, to=8,label='Cylinder', variable=Cylinders_var, orient=HORIZONTAL, length=300)
w_Cylinder.pack()

w_Displacement=Scale(master, from_=0, to=800,label='Displacement', variable=Displacement_var, orient=HORIZONTAL, length=300)
w_Displacement.pack()

w_Horsepower=Scale(master, from_=25, to=300,label='Horsepower', variable=Horsepower_var, orient=HORIZONTAL, length=300)
w_Horsepower.pack()

w_Weight=Scale(master, from_=500, to=7000,label='Weight', variable=Weight_var, orient=HORIZONTAL, length=300)
w_Weight.pack()

w_Acceleration=Scale(master, from_=1, to=25,label='Acceleration', variable=Acceleration_var, orient=HORIZONTAL, length=300)
w_Acceleration.pack()

w_Year=Scale(master, from_=1970, to=2020,label='Year', variable=Year_var, orient=HORIZONTAL, length=300)
w_Year.pack()

R_Europe = Radiobutton(master, text="Europe", variable=Radio_var, value=1)
R_Europe.select()
R_Europe.pack( anchor = W )

R_US = Radiobutton(master, text="US", variable=Radio_var, value=2)
R_US.pack( anchor = W )

R_Japan = Radiobutton(master, text="Japan", variable=Radio_var, value=3)
R_Japan.pack( anchor = W)

w_MPG=Scale(master, from_=0, to=100,label='MPG', variable=MPG_var,resolution=0.01, orient=HORIZONTAL, length=300)
w_MPG.set(w_Cylinder.get()*coeff['Cylinders']
            +w_Displacement.get()*coeff['Displacement']
            +w_Horsepower.get()*coeff['Horsepower']
            +w_Weight.get()*coeff['Weight']
            +w_Acceleration.get()*coeff['Acceleration']
            +(w_Year.get()-1900)*coeff['Year']
            +coeff['Intercept']
         )
w_MPG.pack()




Cylinders_var.trace("w", trace_method)
Displacement_var.trace("w", trace_method)
Horsepower_var.trace("w", trace_method)
Weight_var.trace("w", trace_method)
Acceleration_var.trace("w", trace_method)
Year_var.trace("w", trace_method)
MPG_var.trace("w", trace_method)
Radio_var.trace("w", trace_method)

master.mainloop()
```
